
``` CodeMirror-line
    survival={}
```

``` CodeMirror-line
    not_survival={}
```

``` CodeMirror-line
    rates={}
```

``` CodeMirror-line
        survival[i]=0
```

``` CodeMirror-line
        not_survival[i]=0
```

``` CodeMirror-line
    for i in range(len(train)):
```

``` CodeMirror-line
        if train['Survived'][i]==1:
```

``` CodeMirror-line
            not_survival[train[feature_name][i]]+=1
```

``` CodeMirror-line
    
```

``` CodeMirror-line
    for i in categories:
```

``` CodeMirror-line
        rates[i]=survival[i]*100/(survival[i]+not_survival[i])
```

``` CodeMirror-line
    
```

``` CodeMirror-line
    return rates
```
